```python
import os
import traceback
from pathlib import Path

import application_utils as ap


def make_dir(File_Path):
    if not os.path.exists(File_Path):
        # 目录不存在，进行创建操作
        os.makedirs(File_Path)  # 使用os.makedirs()方法创建多层目录


if __name__ == "__main__":
    holedataset = ap.RGBPReader()
    # rgb_path_list = ['/data/4TSSD/cby/dataset/kitti_depth/kitti_raw_unzipped',
    #                  '/data/4TSSD/cby/dataset/NYUv2/NYUv2_img']
    rgb_path_list = ['/data/4TSSD/cby/dataset/NYUv2/NYUv2_img', ]
    try:
        for rgb_path in rgb_path_list:
            rgb_path = Path(rgb_path)
            for file in rgb_path.rglob('*.png'):
                str_file = str(file)
                print(str_file)
                if 'kitti' in str_file and 'image_02' in str_file:
                    point_path = str_file[:35] + \
                        'kitti_dc/data_depth_annotated/train/' + \
                        str_file[65:92] + \
                        'proj_depth/groundtruth/image_02/' + \
                        str_file[-14:]
                    if os.path.exists(point_path):
                        save_dir = point_path.replace(
                            'kitti_dc/data_depth_annotated/train', 'kitti_hole')
                        save_dir = save_dir.replace(
                            'proj_depth/groundtruth/image_02/', '')
                        save_dir_2 = save_dir[:-15]
                        make_dir(save_dir_2)
                        print(f'Now Dealing with {str_file}')
                        holedataset.get_hole(
                            str_file, point_path, save_dir)
                elif 'NYUv2' in str_file:
                    point_path = str_file.replace(
                        'NYUv2_img', 'NYUv2_depth')
                    save_dir = point_path.replace(
                        'NYUv2_depth', 'NYUv2_hole')
                    save_dir_2 = save_dir[:len('/data/4TSSD/cby/dataset/NYUv2/NYUv2_hole/')]
                    print(f'save_dir_2 = {save_dir_2}')
                    make_dir(save_dir_2)
                    print(f'Now Dealing with {str_file}')
                    holedataset.get_hole(
                        str_file, point_path, save_dir)
    except Exception:
        traceback.print_exc()
```

```python
import os 
def get_gpu_info():
    gpu_status = os.popen('nvidia-smi | grep %').read().split('|')[1:]
    gpu_dict = dict()
    for i in range(len(gpu_status) // 4):
        index = i * 4
        gpu_state = str(gpu_status[index].split('   ')[2].strip())
        gpu_power = int(gpu_status[index].split('   ')[-1].split('/')[0].split('W')[0].strip())
        gpu_memory = int(gpu_status[index + 1].split('/')[0].split('M')[0].strip())
        gpu_dict[i] = (gpu_state, gpu_power, gpu_memory)
    return gpu_dict
gpu_dict = get_gpu_info()
print(gpu_dict)
```

```python
from os import defpath
from torch import Tensor
from pathlib import Path
import numpy as np
from PIL import Image
import torch
def depth_read(filename: Path) -> Tensor:
    str_file = str(filename)
    if '.png' in str_file:
        data = Image.open(filename)
        if 'kitti' in str_file and 'vkitti' not in str_file:
            depth = (np.array(data) / 25500.).astype(np.float32)
            depth = np.clip(depth, 0, 1)
        elif 'NYU' in str_file:
            depth = (np.array(data) / 255.).astype(np.float32)
            depth = np.clip(depth, 0, 1)
        if 'HRWSI' in str_file:
            depth = (np.array(data) / 255.).astype(np.float32)
            depth = np.clip(depth, 0, 1)
        if 'vkitti' in str_file:
            depth = (np.array(data) / 65535.).astype(np.float32)
        depth = torch.from_numpy(depth).unsqueeze(0)
        data.close()
    elif '.npy' in str_file:
        if 'DIODE' in str_file:
            dm = np.load(str_file).squeeze()
            dm_mask_fname = str_file.replace('.npy', '_mask.npy')
            dm_mask = np.load(dm_mask_fname)
            # dm = dm.reshape((768, 1024))
            dm = dm * dm_mask
            if 'outdoor' in str_file:
                depth = (dm / 100.).astype(np.float32)
                depth = np.clip(depth, 0, 1)
            elif 'indoors' in str_file:
                depth = (dm / 20.).astype(np.float32)
                depth = np.clip(depth, 0, 1)

        depth = torch.from_numpy(depth).unsqueeze(0)
    else:
        print('Depth reading error! Unseen Type')
    return depth

str_file = '/data/4TSSD/cby/dataset/g2_dataset/vkitti/vkitti_depth/Scene01/15-deg-left/frames/depth/Camera_0/depth_00443.png'
depth = depth_read(Path(str_file)).numpy()
# depth = depth_read(str_file)
print(np.max(depth))
```
