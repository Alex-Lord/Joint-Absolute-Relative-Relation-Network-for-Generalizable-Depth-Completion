```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

# 
depth_image = np.zeros((320, 320))
center_x, center_y = 160, 160  # 

# 
num_lines = 64  # 16
num_lines = int(num_lines / 2) 
max_range = 100  # 
angle_increment = np.pi / 2 / num_lines  # 

# 
radar_image = np.zeros((320, 320))

# 
fig, ax = plt.subplots()

for i in range(num_lines):
    angle = i * angle_increment - np.pi / 2  # -45 ~ 45
    a = np.cos(angle) * 0.0002 # 
    # print(a)
    flag = 0
    for x in range(320):
        y = int(a * (x - center_x)**2 + center_y) + int(160 / num_lines/ 1.8 * i + np.floor(np.power(i, 1.2)))
        # y = int(a * (x - center_x)**2 + center_y) + int(160 / num_lines * i)
        if 0 <= y < 320:
            # print(f'({y}, {x})')
            flag = 1
            depth = 1.0
            if x % 4 == 0:
                radar_image[y, x] = depth
                radar_image[-y, x] = depth
        else:
            print(f'{i}y {y}')
    if flag == 0:
        print(f'{i}')
    # 
    ax.imshow(radar_image, cmap='gray', extent=[0, 320, 0, 320], origin='lower')
    ax.set_title('Simulated Radar Image')
    display(fig)
    clear_output(wait=True)
    time.sleep(0.001)  # 0.1

# colorbar
cbar = fig.colorbar(ax.imshow(radar_image, cmap='gray', extent=[0, 320, 0, 320], origin='lower'))
cbar.remove()  # colorbar
plt.show()

```
