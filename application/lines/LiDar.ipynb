```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

# 生成深度图像（示例）
depth_image = np.zeros((320, 320))
center_x, center_y = 160, 160  # 中心点坐标

# 模拟雷达线
num_lines = 64  # 生成16条雷达线
num_lines = int(num_lines / 2) 
max_range = 100  # 最大检测距离
angle_increment = np.pi / 2 / num_lines  # 角度增量

# 创建雷达图
radar_image = np.zeros((320, 320))

# 创建图形窗口
fig, ax = plt.subplots()

for i in range(num_lines):
    angle = i * angle_increment - np.pi / 2  # 角度从-45 ~ 45
    a = np.cos(angle) * 0.0002 # 抛物线的参数
    # print(a)
    flag = 0
    for x in range(320):
        y = int(a * (x - center_x)**2 + center_y) + int(160 / num_lines/ 1.8 * i + np.floor(np.power(i, 1.2)))
        # y = int(a * (x - center_x)**2 + center_y) + int(160 / num_lines * i)
        if 0 <= y < 320:
            # print(f'({y}, {x})')
            flag = 1
            depth = 1.0
            if x % 4 == 0:
                radar_image[y, x] = depth
                radar_image[-y, x] = depth
        else:
            print(f'第{i}条线有y {y}')
    if flag == 0:
        print(f'第{i}条线全部没取到！')
    # 绘制雷达图
    ax.imshow(radar_image, cmap='gray', extent=[0, 320, 0, 320], origin='lower')
    ax.set_title('Simulated Radar Image')
    display(fig)
    clear_output(wait=True)
    time.sleep(0.001)  # 暂停0.1秒

# 添加colorbar到最终图
cbar = fig.colorbar(ax.imshow(radar_image, cmap='gray', extent=[0, 320, 0, 320], origin='lower'))
cbar.remove()  # 移除colorbar
plt.show()

```
